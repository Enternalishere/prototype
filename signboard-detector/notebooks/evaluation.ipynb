```python
import json
from pathlib import Path
import numpy as np

def iou(box_a, box_b):
    xa1, ya1, wa, ha = box_a
    xa2, ya2 = xa1 + wa, ya1 + ha
    xb1, yb1, wb, hb = box_b
    xb2, yb2 = xb1 + wb, yb1 + hb
    xi1, yi1 = max(xa1, xb1), max(ya1, yb1)
    xi2, yi2 = min(xa2, xb2), min(ya2, yb2)
    inter_area = max(0, xi2 - xi1) * max(0, yi2 - yi1)
    union_area = wa * ha + wb * hb - inter_area
    return inter_area / union_area if union_area > 0 else 0

# Example: load detector output from sample_outputs and compare to labeled boxes.
# ground_truth = { 'mithuna_solutions': [[x, y, w, h], ...] }
# predicted = ...
# compute precision, recall, average IoU.
```
